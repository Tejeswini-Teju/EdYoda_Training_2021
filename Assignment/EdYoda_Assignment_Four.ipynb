### Q 2 ) Write a Python program to triple all numbers of a given list of integers. Use Python map.
#### sample list: [1, 2, 3, 4, 5, 6, 7]

```python
nums = (1, 2, 3, 4, 5, 6, 7) 
print("Example List:\n",nums)
result = map(lambda a: a + a + a, nums) 
print("\nTriple of list")
```

### Q3. Write a Python program to square the elements of a list using map() function.

```python
lst = [4, 5, 2, 9]
result= list(map(lambda l: l**2, lst))
print('Square the given number is:  ', result)
```
