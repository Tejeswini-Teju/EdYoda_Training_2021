### Write a Python class to implement pow(x, n)
Explanation: Use should be able to find the nth power of the x.(i.e xxx*x...n times).
You must implement it using Class Sample Input: x: 10, n: 2

Sample Output: 100

```python
def nth_power(a, b):
    if (b == 0): return 1
    elif (int(b % 2) == 0):
        return (nth_power(a, int(b / 2)) *nth_power(a, int(b / 2)))
    else:
        return (a * nth_power(a, int(b / 2)) *  nth_power(a, int(b / 2)))
a = 10;b = 2
print(nth_power(a,b))
```

```python
#Another Way
class power_of_nth:
    def pow(self, a, n):
        if a==0 or a==1 or n==1:
            return a

        if a==-1:
            if n%2 ==0:
                return 1
            else:
                return -1
        if n==0:
            return 1
        if n<0:
            return 1/self.pow(a,-n)
        val = self.pow(a,n//2)
        if n%2 ==0:
            return val*val
        return val*val*a
    
print( power_of_nth().pow(int(input('a:')), int(input('n:'))))
```
