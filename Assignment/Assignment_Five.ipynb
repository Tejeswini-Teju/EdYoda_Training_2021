### Write a Python class to implement pow(x, n)
Explanation:
Use should be able to find the nth power of the x.(i.e x*x*x*x...n times).You must implement it using Class
Sample Input:
x: 10, n: 2

Sample Output:
100

```python
def nth_power(x, y):
  
    if (y == 0): return 1
    elif (int(y % 2) == 0):
        return (nth_power(x, int(y / 2)) *nth_power(x, int(y / 2)))
    else:
        return (x * nth_power(x, int(y / 2)) *  nth_power(x, int(y / 2)))
x = 10;y = 2
print(nth_power(x,y))
```
